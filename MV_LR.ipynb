```
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
import numpy as np
```

```
df = pd.read_csv("https://raw.githubusercontent.com/mahesh147/Multiple-Linear-Regression/master/50_Startups.csv",sep=',')
pd.DataFrame(df)
```

```
x = df[['R&D Spend','Administration','Marketing Spend']]
y = df['Profit']
for i in range(0.1, 0.5, 0.05)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```
rr = Lasso()
rr.fit(x_train, y_train)
yp = rr.predict(x_test)
```

```
plt.subplot(2,2,1)
plt.title("Variation of Profit with R&D")
plt.scatter(x['R&D Spend'],y,color = 'r')
plt.xlabel("R&D spend")
plt.ylabel("profit")
plt.show()

plt.subplot(2,2,2)
plt.title("Variation of Profit with Administration")
plt.scatter(x['Administration'],y,color = 'b')
plt.xlabel("Administration")
plt.ylabel("profit")
plt.show()

plt.subplot(2,2,3)
plt.title("Variation of Profit with Marketing Spend")
plt.scatter(x['Marketing Spend'],y,color = 'orange')
plt.xlabel("Marketing Spend")
plt.ylabel("profit")
plt.show()
```

```
print("The accuracy of this model is",str(100*lr.score(x_train,y_train)),"% on training data", "and", str(100*lr.score(x_test,y_test)),"% on predicted data")
```
